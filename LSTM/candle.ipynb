```python
import pandas as pd
import numpy as np
import talib as ta
from pandas_datareader import data
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.simplefilter('ignore')
import mplfinance as mpf
```

```python
start = '2019-07-01'
end = '2020-07-01'

df = data.DataReader('^N225', 'yahoo', start, end)
```

```python
date = df.index
close = df['Adj Close']
```

```python
span5 = 5
span25 = 25
span50 = 50
df['upper'], df['middle'], df['lower'] = ta.BBANDS(close, timeperiod=span25, nbdevup=2, nbdevdn=2, matype=0)
```

```python
df_candle = df[['High', 'Low', 'Open', 'Close', 'Volume']]
```

```python
mpf.plot(df_candle, type='candle', style='yahoo', figsize=(30, 15))
```

```python
mpf.plot(df_candle[0:50], type='candle', style='yahoo', figsize=(30, 15))
```

```python
mpf.plot(df_candle[100:200], type='candle', style='yahoo', figsize=(30, 15))
```

```python
tcdf = df[['upper', 'middle', 'lower']]
apd = mpf.make_addplot(tcdf)

mpf.plot(df_candle, addplot=apd, type='candle', volume=True, style='yahoo', figsize=(30, 15))
```
